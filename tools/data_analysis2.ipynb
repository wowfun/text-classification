```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
train_ds='../data/train/labeled_data_processed.csv'
# test_ds='../data/test_data.csv'

train_df=pd.read_csv(train_ds)[['cuted_content','label']]
train_df
```

```python
train_df['word_count']=train_df['cuted_content'].str.count(' ')+1
print(train_df['word_count'].describe())

# 词数分布直方图
bins=int((train_df['word_count'].max()-train_df['word_count'].min())**0.5)
print(bins)
plt.hist(train_df['word_count'],bins=bins)
plt.xlabel('Word count')
plt.title('Word count')
# 结论： 绝大部分文本
```
